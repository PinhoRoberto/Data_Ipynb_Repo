```python
import numpy as np

lista_python = [1, 2, 3, 4, 5]
array_numpy = np.array([1, 2, 3, 4, 5])

print("Lista Python:", lista_python)
print("Array NumPy:", array_numpy)
print("Tipo:", type(array_numpy))
```

```python
my_list = [10, 20, 30, 40, 50]
double_list = [x * 2 for x in my_list]

my_array = np.array([1,2,3,4,5])
double_array = my_array * 2

print("Doble Lista Python:", double_list)
print("Doble Array NumPy:", double_array)
```

```python
zeros = np.zeros(5)
ones = np.ones(5)
sequence = np.arange(0, 10, 2)
random_nums = np.random.randint(1, 100, 5)

print("Zeros:", zeros)
print("Ones:", ones)
print("Sequence:", sequence)
print("Random:", random_nums)
```

```python
number = np.random.randint(1, 100, 10)
np.sum(number)
np.mean(number)
np.std(number)
np.min(number)
np.max(number)
print("Random Number:", number)
print("Sum of Random Numbers:", np.sum(number))
print("Mean of Random Numbers:", np.mean(number))
print("Standard Deviation of Random Numbers:", np.std(number))  
print("Min of Random Numbers:", np.min(number))
print("Max of Random Numbers:", np.max(number))
```
