```python
import numpy as np
import time

```

```python
size = 1_000_000

python_list = list(range(size))
start = time.time()
python_result = [x*2 for x in python_list]
python_time = time.time() - start

num_array = np.arange(size)
start = time.time()
num_result = num_array * 2
num_time = time.time() - start

print(f"Python list: {python_time:.5f} seconds")
print(f"Numpy time: {num_time:.5f} seconds")
print(f"Numpy is {python_time/num_time:.2f} times faster than Python list")
```
