```python
import pandas as pd
```

```python
truck = {
    'maker': 'Ford',
    'model': 'F-150',
    'year': 2020,
    'color': 'blue',
    'price': '$35000'
}
car = {
    'maker': 'Toyota',
    'model': 'Camry',
    'year': 2021,
    'color': 'red',
    'price': '$24000'
}

truck_series = pd.Series(truck)
car_series = pd.Series(car)

print("Truck Series:")
print(truck_series)
print("\nCar Series:")
print(car_series)
```
