```python
import numpy as np

numbers = np.random.randint(1, 100, 10)

total = np.sum(numbers)
mean = np.mean(numbers)
min_val = np.min(numbers)
max_val = np.max(numbers)
max_index = np.argmax(numbers)

above_mean = numbers[numbers > mean]

print("Random array:", numbers)
print("-" * 40)
print(f"Sum: {total}")
print(f"Mean: {mean:.2f}")
print(f"Min: {min_val} | Max: {max_val}")
print(f"Max value is at index: {max_index}")
print("-" * 40)
print(f"Values above mean: {above_mean}")
print(f"Count above mean: {len(above_mean)}")
```
