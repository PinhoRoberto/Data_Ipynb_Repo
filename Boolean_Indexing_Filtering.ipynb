```python
temperatures = np.array([22.5, 25.0, 19.0, 30.5, 27.0, 15.5])

hot_days = temperatures[temperatures > 30.0]
print(f"hot_days (>30.0°C): {hot_days}")

comfortable = temperatures[(temperatures >= 20.0) & (temperatures <= 30.0)]
print(f"comfortable (20.0-25.0°C): {comfortable}")

cold_days = temperatures[(temperatures < 20.0) | (temperatures > 35.0)]
print(f"cold_days (<20.0°C or >35.0°C): {cold_days}")
```
